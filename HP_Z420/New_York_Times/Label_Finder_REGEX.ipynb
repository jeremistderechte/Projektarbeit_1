```python
import pickle
import pandas as pd
import string
import re

data = pd.read_csv("./new_york_times_cleanend.csv")["commentBody"]

with open('index', "rb") as f:
    indexList = pickle.load(f)
    
with open('label', "rb") as f:
    labelList = pickle.load(f)
    
dataframeLabelTimes = pd.DataFrame(columns=["sentence", "labels"])    
    

sentenceList = []
for index in indexList:
    sentenceList.append(data.iloc[index])

dataframeLabelTimes["sentence"] = sentenceList
dataframeLabelTimes["labels"] = labelList       
```

```python
len(dataframeLabelTimes["labels"])
```

```python
sentenceList = []
rangeList = []

for label, sentence in zip(dataframeLabelTimes["labels"], dataframeLabelTimes["sentence"]):
    tempList = []
    tempList = label.split(", ")
    
    sentence = re.sub("\^", "^ ", str(sentence))
    sentence = re.sub("-", " - ", str(sentence))
    sentence = re.sub("\~", "~ ", str(sentence))    
    sentence = ' '.join(sentence.split())
    
    
    
    if not (tempList[0] == ""):
        outerRange = []
        sentenceList.append(sentence)
        for name in tempList:
            for match in re.finditer(name, sentence):
                tempRange = []
                tempRange.append(match.start())
                tempRange.append(match.end())
                outerRange.append(tempRange)      
        rangeList.append(outerRange)

    else:
        sentenceList.append(sentence)
        rangeList.append("")

        
```

```python
dataframeLabelTimes["sentence"] = sentenceList
dataframeLabelTimes["ranges"] = rangeList
```

```python
dataframeLabelTimes.to_csv("New_York_Times_labeled_dataframe.csv")
```
